# Describing Distributions

## Importing Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from scipy import stats
```

```python
df = pd.read_csv(r"C:\Users\aliss\Data\Medical Dataset.csv")
df.info()
```

##  Histogram w/KDE

### Doctor Visits

```python
mean = df['Doc_visits'].mean()
std = df['Doc_visits'].std()
plus_one_std = mean + std
minus_one_std = mean - std
```

```python
ax = sns.histplot(x=df['Doc_visits'],stat='probability',kde=True)
ax.axvline(df['Doc_visits'].mean(), color='r')
ax.axvline(df['Doc_visits'].std(), color='g')
ax.axvline(plus_one_std, color = 'black')
ax.axvline(minus_one_std, color = 'black')
ax.axvspan(plus_one_std, minus_one_std, color = 'yellow', zorder = 0)
ax.set_title("Doctor Visits");
```

#### Answering Questions

##### Is it Discrete or Continuous?

This data is discrete

##### Does it have a skew? If so, which direction (+/-)

From what I'm seeing when I look at the histagram, there does not appear to be a skew as it does look to be about even on each side of the mean line 

##### What type of kurtosis does it display? (Mesokurtic, Leptokurtic, Platykurtic)

```python
kurt = stats.kurtosis(df['Doc_visits'], fisher = False)
kurt
```

Mesokurtic
